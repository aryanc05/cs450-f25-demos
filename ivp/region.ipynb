```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = np.meshgrid(np.linspace(-3,3, 1000),np.linspace(-3,3, 1000))
```

```python
Z = np.abs(1 + X + 1j * Y
           + (1/2) * (X + 1j * Y)**2
          )#+ (1/6) * (X + 1j * Y)**3 + (1/24) * (X + 1j * Y)**4)
I, J = np.where(Z <=1)
Z[I,J] = 0
I, J = np.where(Z > 1)
Z[I,J] = 1
fig, ax = plt.subplots()
ax.set_aspect(1.0)
plt.pcolormesh(X, Y, Z)
```
