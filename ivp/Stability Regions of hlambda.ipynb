```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = np.meshgrid(np.linspace(-3,3,100), np.linspace(-3,3,100))
# h \lambda = X + i Y 
```

```python
hlambda = X + 1j * Y
Z = np.abs(1 + hlambda + (1/2) * hlambda**2 + (1/6) * hlambda**3 + (1/24) * hlambda**4)
IJ = np.where(Z <= 1)
Z[IJ] = 0
IJ = np.where(Z > 1)
Z[IJ] = 1
fig, ax = plt.subplots(figsize=(3,3))
ax.set_aspect(1.)
p = ax.pcolormesh(X, Y, Z)
plt.colorbar(p)
```
