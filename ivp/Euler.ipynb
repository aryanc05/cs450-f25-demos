```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
# y' = -100 y = alpha y
alpha = -100

h = 0.009
T = 0.5
nsteps = int(T/h)
y = np.zeros(nsteps)
y[0] = 1

for k in range(1,nsteps):
    y[k] = y[k-1] + h * (alpha * y[k-1])
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.plot([k*h for k in range(nsteps)], y, 'o-', ms=3, mfc='w')
```

```python
h = 10
T = 20
nsteps = int(T/h)
y = np.zeros(nsteps)
y[0] = 1

def g(y1, h, y0):
    """
    minimize g(y1) with parameters h and y0
    """
    return y1 + h*y1**3 - y0

def dg(y1, h, y0):
    return 1 + 3 * h * y1**2
    
for k in range(1,nsteps):
    y[k] = y[k-1] + h * (-y[k-1]**3)
    y[k] = opt.newton(g,     # function to find the root of
                      y[k],  # initial guess
                      dg,
                      args=(h, y[k-1]))
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.plot([k*h for k in range(nsteps)], y, 'o-', ms=3, mfc='w')
```
